```python
import pandas as pd
import os
```

```python
path = os.path.abspath(os.path.join('..', 'input'))
path_data = os.path.join(path, 'WWBICountry.csv')
df = pd.read_csv(path_data)
```

```python
df.info()
```

```python
df_colunas = df[["Short Name",
"Country Code",
"Currency Unit",
"Special Notes",
"Region",
"Income Group",
"National accounts base year",
"National accounts reference year",
"Lending category",
"Other groups",
"SNA price valuation",
"System of National Accounts",
"IMF data dissemination standard",
"System of trade",
"Balance of Payments Manual in use",
"External debt Reporting status",
"Government Accounting concept",
"Latest population census",
"Latest agricultural census",
"Latest household survey",
"Latest industrial data",
"Source of most recent Income and expenditure data",
"Vital registration complete"]].copy()
```

```python
df_colunas.rename(columns={"Short Name":"country_name",
"Country Code":"country_code",
"Currency Unit":"currency_unit",
"Special Notes":"special_note",
"Region":"region",
"Income Group":"income_group",
"National accounts base year":"national_accounts_base_year",
"National accounts reference year":"national_accounts_reference_year",
"Lending category":"lending_category",
"Other groups":"other_groups",
"SNA price valuation":"sna_price_valuation",
"System of National Accounts":"sna_methodology",
"IMF data dissemination standard":"imf_dissemination_standard",
"System of trade":"trade_system",
"Balance of Payments Manual in use":"balance_payments_manual",
"External debt Reporting status":"external_debt_reporting",
"Government Accounting concept":"government_accounting",
"Latest population census":"population_census",
"Latest agricultural census":"agricultural_census",
"Latest household survey":"household_survey",
"Latest industrial data":"industrial_date",
"Source of most recent Income and expenditure data":"source_income_expenditure",
"Vital registration complete":"vital_registration_complete"},inplace = True)
```

```python
df_colunas
```

```python
df_colunas['industrial_date'] = df_colunas['industrial_date'].astype('Int64')
df_colunas['national_accounts_reference_year'] = df_colunas['national_accounts_reference_year'].astype('Int64')
df_colunas['vital_registration_complete'].replace('Yes', 1, inplace = True)
df_colunas['vital_registration_complete'] = df_colunas['vital_registration_complete'].astype('Int64')
```

```python
path = os.path.abspath(os.path.join('..', 'output'))
path_df = os.path.join(path, 'country.csv')
df_colunas.to_csv(path_df, index=False)
```
