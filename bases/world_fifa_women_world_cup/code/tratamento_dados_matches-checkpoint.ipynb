### Transformando algumas variáveis 


```python
df_dados[["team_a_score_penalties", "team_b_score_penalties"]] = df_dados[["team_a_score_penalties","team_b_score_penalties"]].fillna(0)
df_dados["match_id"] = df_dados["match_id"].astype(str)
df_dados[["year","team_a_score", "team_b_score","team_a_score_penalties", "team_b_score_penalties"]] = df_dados[["year","team_a_score", "team_b_score","team_a_score_penalties", "team_b_score_penalties"]].astype(int)
df_dados["team_a_score_margin"] = df_dados["team_a_score"] - df_dados["team_b_score"]
df_dados["team_b_score_margin"] = df_dados["team_b_score"] - df_dados["team_a_score"]
```

### Achando resultados 

```python
def label(data):
    if data["team_a_score"] < data["team_b_score"]: 
        return "team_b_win"
    elif data["team_b_score"] < data["team_a_score"]:
        return "team_a_win" 
    elif data["team_a_score_penalties"] < data["team_b_score_penalties"]:
        return "team_b_win"
    elif data["team_b_score_penalties"] < data["team_a_score_penalties"]:
        return "team_a_win"
    elif data["team_b_score"] == data["team_a_score"]:
        return "draw"
```

```python
df_dados["win"] = df_dados.apply(lambda df_dados:label(df_dados),axis=1)
df_dados
```
