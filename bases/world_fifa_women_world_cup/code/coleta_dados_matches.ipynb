```python
def stage(match):
    if  match <= len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) :
        return "Groups"
    elif (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) ) < match <= (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 4):
        return "Quarter-final"
    elif (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 4 ) < match <= (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 6):
        return "Semi-final"
    elif (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 6 ) < match <= (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 7):
        return "Match for third place"
    elif (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 7 ) < match <= (len(df_dados_1991[df_dados_1991["stage_name"] == "Groups"]) + 8):
        return "Final"
```

```python
df_dados_1991["match_id"] = df_dados_1991["match_id"].astype(int)
```

```python
df_dados_1991["stage_name"] = df_dados_1991["match_id"].apply(lambda x: stage(x))
```

```python
def stage(match):
    if  match <= len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) :
        return "Groups"
    elif (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) ) < match <= (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 8):
        return "Round of 16"
    elif (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 8 ) < match <= (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 12):
        return "Quarter-final"
    elif (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 12 ) < match <= (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 14):
        return "Semi-final"
    elif (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 14 ) < match <= (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 15):
        return "Match for third place"
    elif (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 15 ) < match <= (len(df_dados_2019[df_dados_2019["stage_name"] == "Groups"]) + 16):
        return "Final"
```

```python
df_dados_2019["match_id"] = df_dados_2019["match_id"].astype(int)
df_dados_2019["stage_name"] = df_dados_2019["match_id"].apply(lambda x: stage(x))
df_dados_2019
```
