```python
mun = geobr.read_municipality()
```

```python
mun['code_muni'] = mun['code_muni'].astype(np.int64)
```

```python
mun
```

```python
#Alterando nomes das colunas pra estilo da BD
mun=mun.rename({'code_muni':'id_municipio', 'code_state':'id_uf', 'abbrev_state':'sigla_uf','geometry':'geometria'},axis='columns')
```

```python
#selecionar variáveis de interesse
mun=mun.filter(items=['id_municipio', 'sigla_uf', 'geometria'])
```

```python
from posixpath import sep
#salvando em csv
mun.to_csv('municipio.csv', sep=',', index=False, header=True)
```
