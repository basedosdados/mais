```python
import geobr
import pandas as pd
import numpy as np
```

```python
saude = geobr.read_health_region(year=2013)
```

```python
saude
```

```python
saude['code_state'] = saude['code_state'].astype(np.int64)
saude['code_health_region'] = saude['code_health_region'].astype(np.int64)
```

```python
saude=saude.rename({'code_health_region':'id_regiao_saude', 'name_health_region':'nome_regiao_saude', 'code_state':'id_uf','geometry':'geometria', 'abbrev_state':'sigla_uf'}, axis='columns')
saude
```

```python
saude = saude.filter(items=['id_regiao_saude', 'id_uf', 'sigla_uf','geometria'])
```

```python
saude
```

```python
saude.to_csv('saude.csv', sep=',', index=False, header='True')
```
