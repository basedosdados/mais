### educação

```python
path_munic = os.path.join(path, 'MUNIC_2018_Educação.csv')
munic = pd.read_csv(path_munic)
munic_educ = munic[['Cod Municipio','MEDU01','MEDU03','MEDU04','MEDU05','MEDU051','MEDU06','MEDU07']]
munic_educ['ano'] = 2018
munic_educ['tema'] = 'educacao'

def lab(munic_educ):
    if (munic_educ['MEDU01'] == 'Recusa'):
        return 1
    elif (munic_educ['MEDU01'] != 'Recusa'):
        return 0
munic_educ["recusa"] = munic_educ.apply(lambda munic_educ:lab(munic_educ),axis = 1)

def lab2(munic_educ):
    if (munic_educ['MEDU03'] == '(**) Sem gestor'):
        return 1
    elif (munic_educ['MEDU03'] != '(**) Sem gestor'):
        return 0
munic_educ["sem_gestor"] = munic_educ.apply(lambda munic_educ:lab2(munic_educ),axis = 1)

munic_educ.rename(columns = {'Cod Municipio':'id_municipio', 
                             'MEDU01':'caracterizacao_orgao_gestor',
                             'MEDU03':'sexo',
                             'MEDU04':'idade', 
                             'MEDU05':'cor_raca',
                             'MEDU051':'cor_raca_autodeclarado',
                             'MEDU06':'escolaridade',
                             'MEDU07':'escolaridade_formacao'}, inplace = True)
munic_educ = munic_educ[['ano',
                         'id_municipio',
                         'tema',
                         'recusa',
                         "sem_gestor",
                         'caracterizacao_orgao_gestor',
                         'sexo',
                         'idade',
                         'cor_raca',
                         'cor_raca_autodeclarado',
                         'escolaridade',
                         'escolaridade_formacao']]
munic_educ.replace('-', np.nan, inplace = True)
munic_educ.replace('Recusa', np.nan, inplace = True)
munic_educ.replace('(*) Não soube informar', np.nan, inplace = True)
munic_educ['cor_raca_autodeclarado'].replace('Sim', 1, inplace = True)
munic_educ['cor_raca_autodeclarado'].replace('Não', 0, inplace = True)
munic_educ['cor_raca_autodeclarado'] = munic_educ['cor_raca_autodeclarado'].astype('float').astype('Int64')
munic_educ['id_municipio'] = munic_educ['id_municipio'].astype(str)
munic_educ['idade'] = munic_educ['idade'].astype('float').astype('Int64')
munic_educ
```
