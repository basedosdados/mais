### recursos humanos

```python
path_munic = os.path.join(path, 'MUNIC_2018_recursos.csv')
munic_recursos = pd.read_csv(path_munic)
```

```python
munic_recursos['ano'] = 2018
```

```python
def lab(munic_recursos):
    if (munic_recursos['MREH02'] == 'Recusa'):
        return 1
    elif (munic_recursos['MREH02'] != 'Recusa'):
        return 0

munic_recursos["recusa"] = munic_recursos.apply(lambda munic_recursos:lab(munic_recursos),axis = 1)
```

### direta

```python
munic_recursos_direta = munic_recursos.copy()
munic_recursos_direta['tipo_administracao'] = 'Direta'
```

```python
munic_recursos_direta['MREH0124'] = 0
```

```python
quant_munic_direta = pd.DataFrame()

for i in ['Estatutários','Celetistas', 'Somente comissionados', 'Estagiários', 'Sem vínculo permanente']:
    munic_recursos_direta['tipo_vinculo'] = i
    for t in ['Total','Sem instrução','Ensino fundamental', 'Ensino médio', 'Ensino superior','Pós-graduação' ]:
        munic_recursos_direta['escolaridade'] = t
        temp1 = munic_recursos_direta[['ano','Cod Municipio','recusa','tipo_administracao','MREH02','tipo_vinculo','escolaridade']]
        quant_munic_direta = quant_munic_direta.append(temp1).reset_index(drop=True)
    
```

```python
quant_munic_direta[(quant_munic_direta['escolaridade'] == 'Sem instrução')&(quant_munic_direta['tipo_vinculo'] == 'Estatutários')]
```

```python
valor_quant_munic_direta = pd.DataFrame()
for j in ['1','2','3','4','5']:
    for i in ['1','2','3','4','5','6']:
        lista = munic_recursos_direta['MREH01'+i+j].values
        lista = pd.DataFrame(lista)
        valor_quant_munic_direta = valor_quant_munic_direta.append(lista).reset_index(drop=True)
```

```python
valor_quant_munic_direta = valor_quant_munic_direta.rename(columns={0: 'quantidade_vinculo'})
```
