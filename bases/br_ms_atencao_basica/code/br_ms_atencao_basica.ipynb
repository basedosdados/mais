```python
import pandas as pd
import numpy as np
import basedosdados as bd
```

```python
all_dfs = pd.read_excel("C:/Users/laura/OneDrive/Documents/BasedeDados/br_ms_atencao_basica/input/br_ms_atencao_basica_bruta.xlsx", 
                        sheet_name=None,
                        decimal = ',',
                        thousands = '.')
```

```python
all_dfs.pop('Informações')
```

```python
dicionario_renomear = { 'NU_COMPETENCIA':'ano',
                        'SG_UF':'sigla_uf',
                        'CO_MUNICIPIO_IBGE':'id_municipio_6',
                        'QT_POPULACAO':'populacao',
                        'QT_CH_MEDICO':'carga_horaria_medica_atencao_basica_tradicional',
                        'QT_CH_ENFERMEIRO':'carga_horaria_enfermagem_atencao_basica_tradicional',
                        'QT_EQUIPE_AB_EQUIVALENTE_CH':'quantidade_equipes_atencao_basica_equivalente',
                        'QT_EQUIPE_AB_PARAMETRIZADA':'quantidade_equipes_atencao_basica_parametrizada',
                        'QT_EQUIPE_SF':'quantidade_equipes_saude_familia',
                        'QT_EQUIPE_SF_AB':'quantidade_equipes_atencao_basica_total',
                        'QT_COBERTURA_SF':'populacao_coberta_estrategia_saude_familia',
                        'PC_COBERTURA_SF':'proporcao_cobertura_estrategia_saude_familia',
                        'QT_COBERTURA_AB':'populacao_coberta_total_atencao_basica',
                        'PC_COBERTURA_AB':'proporcao_cobertura_total_atencao_basica'
                      }
```

```python
colunas_selecionadas = ['ano',
                        'mes',
                        'sigla_uf',
                        'id_municipio',
                        'id_municipio_6',
                        'populacao',
                        'carga_horaria_medica_atencao_basica_tradicional',
                        'carga_horaria_enfermagem_atencao_basica_tradicional',
                        'quantidade_equipes_atencao_basica_equivalente',
                        'quantidade_equipes_atencao_basica_parametrizada',
                        'quantidade_equipes_saude_familia',
                        'quantidade_equipes_atencao_basica_total',
                        'populacao_coberta_estrategia_saude_familia',
                        'proporcao_cobertura_estrategia_saude_familia',
                        'populacao_coberta_total_atencao_basica',
                        'proporcao_cobertura_total_atencao_basica'
                       ]
```

```python
df = pd.concat(all_dfs, ignore_index=True)
```

```python
df.rename(columns = dicionario_renomear, inplace=True)
```

```python
df['mes'] = df['ano'].astype("str").str[-2:].astype('int64')
df['ano'] = df['ano'].astype("str").str[:4].astype('int64')
df['populacao_coberta_total_atencao_basica'] = df['populacao_coberta_total_atencao_basica'].astype('int64')
df['id_municipio_6'] = df['id_municipio_6'].astype('str')
```

```python
df2 = bd.read_table(
    dataset_id='br_bd_diretorios_brasil',
    table_id='municipio',
    billing_project_id="rfid-250503"
)
```

```python
id_mun = df2[['id_municipio','id_municipio_6']]
```

```python
df = pd.merge(df, id_mun, how = 'left', left_on = 'id_municipio_6', right_on = 'id_municipio_6', validate='many_to_one')
```
