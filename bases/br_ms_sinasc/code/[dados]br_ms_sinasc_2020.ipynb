```python
for uf in ufs:
  print(" Tratando {}".format(uf))
  exec("df = read_dbc('/content/gdrive/MyDrive/Dados/Conjuntos/br_ms_sinasc/input/DN{}2020.dbc', encoding='iso-8859-1')".format(uf))
  df.rename(columns=rename_96_19, inplace=True)
  df['data_nascimento'] = pd.to_datetime(df['data_nascimento'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_cadastro'] = pd.to_datetime(df['data_cadastro'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_recebimento'] = pd.to_datetime(df['data_recebimento'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_recebimento_original'] = pd.to_datetime(df['data_recebimento_original'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_nascimento_mae'] = pd.to_datetime(df['data_nascimento_mae'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_ultima_menstruacao'] = pd.to_datetime(df['data_ultima_menstruacao'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_declaracao'] = pd.to_datetime(df['data_declaracao'], format ='%d%m%Y',errors='coerce').dt.date
  df['data_nascimento_mae'] = pd.to_datetime(df['data_nascimento_mae'], format ='%d%m%Y',errors='coerce').dt.date
  df['hora_nascimento'] = pd.to_datetime(df['hora_nascimento'],format='%H%M',errors='coerce').dt.time
  try:
    df = df.sort_values(by=['id_municipio_nascimento'])
    if len(df.iloc[0]['id_municipio_nascimento'])==6:
      df = pd.merge(df, id_mun, how='left', left_on='id_municipio_nascimento', right_on= 'id_municipio_6', validate='many_to_one')
      df.drop(columns=['id_municipio_nascimento', 'id_municipio_6', 'sigla_uf', 'regiao'], inplace=True)
      df.rename(columns={'id_municipio':'id_municipio_nascimento'}, inplace=True)
    else:
        pass
  except:pass
  try:
    df = df.sort_values(by=['id_municipio_mae'])
    if len(df.iloc[0]['id_municipio_mae'])==6:
      df = pd.merge(df, id_mun, how='left', left_on='id_municipio_mae', right_on= 'id_municipio_6', validate='many_to_one')
      df.drop(columns=['id_municipio_mae', 'id_municipio_6', 'sigla_uf', 'regiao'], inplace=True)
      df.rename(columns={'id_municipio':'id_municipio_mae'}, inplace=True)
    else:
      pass
  except:pass
  try:
    df = df.sort_values(by=['id_municipio_residencia'])
    if len(df.iloc[0]['id_municipio_residencia'])==6:
      df = pd.merge(df, id_mun, how='left', left_on='id_municipio_residencia', right_on= 'id_municipio_6', validate='many_to_one')
      df.drop(columns=['id_municipio_residencia', 'id_municipio_6', 'sigla_uf', 'regiao'], inplace=True)
      df.rename(columns={'id_municipio':'id_municipio_residencia'}, inplace=True)
    else:
      pass
  except:pass
  df = pd.merge(df, id_mun[['id_municipio', 'id_municipio_6']], how='left', left_on='id_municipio_mae', right_on= 'id_municipio_6', validate='many_to_one')
  df.drop(columns=['id_municipio_mae', 'id_municipio_6'], inplace=True)
  df.rename(columns={'id_municipio':'id_municipio_mae'}, inplace=True)
  df = pd.concat([df, add_14_19])
  df = df[ordem]  
  exec('df.to_csv("/content/gdrive/MyDrive/Dados/Conjuntos/br_ms_sinasc/output/ano=2020/sigla_uf={}/microdados.csv",index=False, encoding="utf-8", na_rep="", date_format="%Y-%m-%d")'.format(uf))
del df
```
