```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_csv('../input/WDI_csv/WDICountry.csv') # acessa o csv
df_arquitetura = pd.read_excel('../extra/architecture/country.xlsx') # acessa a planilha de arquitetura

# cria uma lista com a lista de nomes antigos e novos
old_columns = df_arquitetura['original_name ->'].to_list()
new_columns = df_arquitetura['name'].to_list()

df = df[old_columns] # define o dataframe conforme a ordem na planilha de arquitetura
df.columns = new_columns # substitui o nome das colunas pelo novo nome
# substitui vírgulas em ponto
df['most_recent_income_and_expenditure_data']=df['most_recent_income_and_expenditure_data'].str.replace(',','.')

# substitui nan por 0 e transforma float em int
df['latest_industrial_data'] = df['latest_industrial_data'].fillna(0)
df['latest_industrial_data'] = df['latest_industrial_data'].astype(int)

# substitui nan por 0 e transforma float em int
df['national_accounts_reference_year'] = df['national_accounts_reference_year'].fillna(0)
df['national_accounts_reference_year'] = df['national_accounts_reference_year'].astype(int)

df.to_csv('../output/country.csv',index=False) # salva o novo dataframe na pasta output

df.head()
```
