```python
import basedosdados as bd
import os
import pandas as pd
```

```python
path = os.path.abspath(os.path.join('..', 'input'))
```

```python
path_tabela = os.path.join(path, 'tabela.xlsx')
df = pd.read_excel(path_tabela)
```

```python
df.rename(columns = {'Ano':'ano',
'Unidade da Federação':'nome',
'Cód.':'id_uf',
'Produto Interno Bruto a preços correntes (Mil Reais)':'pib',
'Impostos, líquidos de subsídios, sobre produtos a preços correntes (Mil Reais)':'impostos_liquidos',
'Valor adicionado bruto a preços correntes total (Mil Reais)':'va',
'Valor adicionado bruto a preços correntes da agropecuária (Mil Reais)':'va_agropecuaria',
'Valor adicionado bruto a preços correntes da indústria (Mil Reais)':'va_industria',
'Valor adicionado bruto a preços correntes dos serviços, exclusive administração, defesa, educação e saúde públicas e seguridade social (Mil Reais)':'va_servicos',
'Valor adicionado bruto a preços correntes da administração, defesa, educação e saúde públicas e seguridade social (Mil Reais)':'va_adespss'}, inplace = True)
```

```python
df
```

```python
uf = bd.read_table(dataset_id='br_bd_diretorios_brasil',
table_id='uf',
billing_project_id="dados-bd-359914")
```

```python
uf = uf[['sigla','nome']]
```

```python
df_sigla = pd.merge(uf, df, on = 'nome')
```

```python
df_sigla.rename(columns = {'sigla':'sigla_uf'},inplace = True)
```

```python
df_sigla = df_sigla[['ano','id_uf','sigla_uf','pib','impostos_liquidos','va','va_agropecuaria', 'va_industria','va_servicos','va_adespss']]
```

```python
df_sigla
```

```python
path = os.path.abspath(os.path.join('..', 'output'))
path_df = os.path.join(path, 'uf.csv')
df_sigla.to_csv(path_df, index=False)
```
